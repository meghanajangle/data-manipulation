## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np
Baltimore = pd.read_csv('baltimore_9-24-2016_9-30-2017.csv')
Boston = pd.read_csv('boston_9-24-2016_9-30-2017.csv')
NewYork = pd.read_csv('new-york_9-24-2016_9-30-2017.csv')
Philadelphia = pd.read_csv('philadelphia_9-24-2016_9-30-2017.csv')
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here
for col in Baltimore.columns:
   pct_missing = np.mean(Baltimore[col].isnull())
   print('{} - {}%'.format(col, round(pct_missing*100)))

# dropping Missing Data
Baltimore_copy = Baltimore.drop(["Type", "Sub Variety", "Grade", "Origin District", "Color", "Environment", "Quality", "Condition", "Appearance","Storage", "Crop","Trans Mode"], axis=1) 

Baltimore_copy['Variety'] = Baltimore_copy['Variety'].fillna("_MISSING_")
Baltimore_copy['Origin'] = Baltimore_copy['Origin'].fillna("_MISSING_")
Baltimore_copy['Item Size'] = Baltimore_copy['Item Size'].fillna("_MISSING_")

#finding outliers
Baltimore_copy.boxplot()

#finding outliers for categorical columns
Baltimore_copy['Variety'].value_counts().plot.bar()

Baltimore_copy['Item Size'].value_counts().plot.bar()

# Dropping Unnecassry Data
Baltimore_copy = Baltimore_copy.drop(["Mostly Low"], axis=1) 

Baltimore_copy.info()
```

```python
for col in Boston.columns:
    pct_missing = np.mean(Boston[col].isnull())
    print ('{} - {}%'.format(col,round (pct_missing*100)))
Boston_copy = Boston.drop(['Type', 'Sub Variety', 'Grade','Origin District', 'Environment', 'Quality','Condition','Appearance','Storage','Crop','Trans Mode'],axis = 1)

Boston_copy['Item Size'] = Boston_copy['Item Size'].fillna('_MISSING_')
Boston_copy['Color'] = Boston_copy['Color'].fillna('_MISSING_')

Boston_copy.info()

#finding outliers
Boston_copy.boxplot()

#finding outliers for categorical columns
Boston_copy['Item Size'].value_counts().plot.bar()

#Dropping Unnecassry Data - No Unncessary data found
#Boston_copy = Boston_copy.drop([""], axis=1) 

Boston_copy.info()
```

```python
for col in NewYork.columns:
    pct_missing = np.mean(NewYork[col].isnull())
    print ('{} - {}%'.format(col,round (pct_missing*100)))

  
NewYork_copy = NewYork.drop(['Type', 'Color', 'Sub Variety', 'Grade','Origin District', 'Environment', 'Quality','Condition','Appearance','Storage','Crop','Trans Mode'],axis = 1)
NewYork_copy['Item Size'] = NewYork_copy['Item Size'].fillna('_MISSING-')


#finding outliers
NewYork_copy.boxplot()

#finding outliers for categorical columns
NewYork_copy['Item Size'].value_counts().plot.bar()


#Dropping Unnecassry Data - mo unnessary data 
#NewYork_copy = NewYork_copy.drop([""], axis=1) 

NewYork_copy.info()
```

```python
# Clean your data here
for col in Philadelphia.columns:
    pct_missing = np.mean(Philadelphia[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

Philadelphia.info()
# dropping Missing Data
Philadelphia_copy = Philadelphia.drop(["Type", "Sub Variety", "Grade", "Origin District", "Color", "Environment", "Quality", "Condition", "Appearance","Storage", "Crop","Trans Mode"], axis=1) 


Philadelphia_copy['Item Size'] = Philadelphia_copy['Item Size'].fillna("_MISSING_")

#finding outliers
Philadelphia_copy.boxplot()

#finding outliers for categorical columns
Philadelphia_copy['Item Size'].value_counts().plot.bar()

# Dropping Unnecassry Data
Philadelphia_copy = Philadelphia_copy.drop(["Mostly High"], axis=1) 

Philadelphia_copy.info()
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
#Combined_df = pd.concat([Baltimore_copy,Boston_copy,NewYork_copy,Philadelphia_copy], axis=1, join='outer')
#display(Combined_df)

northeast_df = Baltimore_copy .append([Boston_copy,NewYork_copy ,Philadelphia_copy])
northeast_df.head()
#Combined_df = pd.merge(Baltimore_copy,Boston_copy,NewYork_copy,Philadelphia_copy,indicator=True)
#print(Combined_df)
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.

northeast_df.groupby('Unit of Sale').agg({'Low Price' : 'mean', 'High Price' : 'mean' } , skipna = True)
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.

northeast_df.groupby(['City Name']).agg({'City Name' : 'count'})
```
